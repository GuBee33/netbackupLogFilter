```python
import pandas as pd
```

```python
def nbjm2df(folder,file_name):
    log_data  = open(folder+"/"+file_name, 'r')
    split_list = []
    for line in log_data:
        if line[:2].isdigit():
            linepart = line.split(' ')
            datetime = " ".join(linepart[:2])
            log = " ".join(linepart[2:])
            split_list.append([datetime, log])

    nbjm_df = pd.DataFrame(split_list, columns=['datetime', 'log'])
    nbjm_df["date"] = pd.to_datetime(nbjm_df.datetime).dt.date
    nbjm_df["time"] = pd.to_datetime(nbjm_df.datetime).dt.time
    nbjm_df["time_r"] = pd.to_datetime(nbjm_df.datetime).dt.round('5min').dt.time
    display(nbjm_df.head())
    display(nbjm_df.info())
    return nbjm_df
```

```python
def log2df(folder,file_name):
    log_data  = open(folder+"/"+file_name, 'r',encoding='unicode_escape')
    split_list = []
    for line in log_data:
        if line[:2].isdigit():
            linepart = line.split(' ')
            time = linepart[0]
            pid = linepart[1]
            loglevel = linepart[2]
            log = " ".join(linepart[3:])
            split_list.append([time, pid,loglevel,log])

    df = pd.DataFrame(split_list, columns=['time','pid','loglevel', 'log'])
    df["time_r"] = pd.to_datetime(df.time).dt.round('5min').dt.time
    df.time=pd.to_datetime(df.time).dt.time
    display(df.head())
    display(df.info())
    return df
```

```python
folder='.'
file_name = "nbjm_log"
nbjm_df=nbjm2df(folder,file_name)
```

```python
import os

dfs={}
folder="logs"
for filename in os.listdir(folder):
    if filename[0]!=".":
        print(filename.split('.')[0])
        dfs[filename.split('.')[0]]=log2df(folder,filename)
```

```python
from datetime import datetime,date,time,timedelta

def FilterDFsByTime(dfs,h,m,deltamin):
    fdfs={}
    filterTime = datetime.combine(date(1,1,1),time(h,m))
    d=timedelta(minutes=deltamin)
    ftmin=(filterTime-d).time()
    ftmax=(filterTime+d).time()
    for logfile in dfs: 
        fdfs[logfile]=dfs[logfile][(dfs[logfile].time>ftmin) & (dfs[logfile].time<ftmax)]
    return fdfs

def FilterDFsByPID(dfs,pid):
    fdfs={}
    for logfile in dfs: 
        fdfs[logfile]=dfs[logfile][dfs[logfile].pid.str.contains(pid)]
    return fdfs

def FilterDFsByStingInLog(dfs,text):
    fdfs={}
    for logfile in dfs: 
        fdfs[logfile]=dfs[logfile][dfs[logfile].log.str.contains(text)]
    return fdfs
```
